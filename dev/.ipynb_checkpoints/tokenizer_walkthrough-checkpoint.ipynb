# Tokenizer Walkthrough

This notebook explains the `nanochat` tokenizer in small steps.

We will:
- build a tiny toy corpus
- train a tiny `RustBPETokenizer`
- inspect the learned token pieces
- see how `nanochat` adds special chat tokens
- connect each toy step back to the real repo files

This notebook is intentionally lightweight: it does not download the pretraining dataset and it does not require a previously trained tokenizer.


````python
from pathlib import Path
import inspect
import os
import sys

repo_root = Path.cwd()
if not ((repo_root / "nanochat").exists() and (repo_root / "pyproject.toml").exists()):
    for candidate in [Path.cwd(), *Path.cwd().parents]:
        if (candidate / "nanochat").exists() and (candidate / "pyproject.toml").exists():
            repo_root = candidate
            break

if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

# Keep notebook artifacts local to the repo unless the user already chose a base dir.
os.environ.setdefault("NANOCHAT_BASE_DIR", str(repo_root / ".cache" / "nanochat"))
Path(os.environ["NANOCHAT_BASE_DIR"]).mkdir(parents=True, exist_ok=True)

missing = []
for module_name in ["rustbpe", "tiktoken", "torch", "tokenizers"]:
    try:
        __import__(module_name)
    except Exception:
        missing.append(module_name)

if missing:
    missing_str = ", ".join(missing)
    raise RuntimeError(
        f"Missing dependencies: {missing_str}\n"
        f"Current Python executable: {sys.executable}\n\n"
        "This usually means Jupyter opened the notebook on the wrong kernel.\n"
        "In Jupyter, switch to the `nanochat (.venv)` kernel and rerun this cell.\n\n"
        "If you have not installed that kernel yet, from the repo root run:\n"
        "  uv sync --extra cpu --group dev\n"
        "  source .venv/bin/activate\n"
        "  python -m ipykernel install --user --name nanochat --display-name 'nanochat (.venv)'"
    )

from nanochat.common import get_base_dir
from nanochat.tokenizer import RustBPETokenizer, SPECIAL_TOKENS, get_tokenizer

print(f"repo_root: {repo_root}")
print(f"python: {sys.executable}")
print(f"base_dir: {get_base_dir()}")
print(f"special token count: {len(SPECIAL_TOKENS)}")
````

## Step 1. Make a tiny corpus

We start with a tiny list of strings so the learned pieces are easy to inspect.

How this maps to `nanochat`:
- In the real repo, [scripts/tok_train.py](../scripts/tok_train.py) does **not** use a toy list.
- It streams many documents from the pretraining dataset through an iterator.
- The idea is the same: feed text into `RustBPETokenizer.train_from_iterator(...)`.


````python
toy_docs = [
    "low",
    "lower",
    "lowest",
    "slow",
    "slower",
    "slowest",
    "new",
    "newer",
    "newest",
]

print("Toy corpus:")
for doc in toy_docs:
    print(f"  - {doc}")
````

## Step 1.5. Understand the nanochat-style text iterator

In the real repo, `scripts/tok_train.py` does **not** pass a Python list directly into tokenizer training.
It builds a generator function called `text_iterator()` that yields one document at a time.

Important Python detail: a function that uses `yield` returns a **generator object** even if you never write an explicit `return`.

How this maps to `nanochat`:
- the real code loops over batches from `parquets_iter_batched(split="train")`
- then loops over each document inside the batch
- optionally crops the document
- `yield`s one `doc_text` at a time
- later, `rustbpe` consumes that generator during tokenizer training

The next cell mirrors that exact shape with a tiny toy version.


````python
toy_batches = [
    toy_docs[:3],
    toy_docs[3:6],
    toy_docs[6:],
]

def toy_text_iterator():
    """A tiny version of the real tok_train.py text_iterator()."""
    nchars = 0
    for batch in toy_batches:               # like: for batch in parquets_iter_batched(...)
        for doc in batch:                   # like: for doc in batch
            doc_text = doc
            if len(doc_text) > 10_000:      # like the doc_cap crop in tok_train.py
                doc_text = doc_text[:10_000]
            nchars += len(doc_text)
            yield doc_text                  # this is what rustbpe consumes later
            if nchars > 1_000_000:          # toy stand-in for max_chars stopping
                return

preview_iter = toy_text_iterator()
print(f"type(preview_iter): {type(preview_iter)}")
print("First yielded doc:", next(preview_iter))
print("Second yielded doc:", next(preview_iter))
print("Remaining docs:", list(preview_iter))
````

## Step 2. Pick a tiny vocabulary budget

`RustBPETokenizer` is byte-based, so it always starts from the 256 possible byte values.
`nanochat` also reserves special chat tokens like `<|bos|>` and `<|user_start|>`.

So even our toy tokenizer needs a vocab size a bit above `256 + len(SPECIAL_TOKENS)`.

How this maps to `nanochat`:
- In the real repo, [scripts/tok_train.py](../scripts/tok_train.py) defaults to `32768`.
- Here we choose something much smaller so only a few merges are learned.


````python
from nanochat.tokenizer import SPECIAL_TOKENS

base_bytes = 256
num_special = len(SPECIAL_TOKENS)
extra_merges = 11

toy_vocab_size = base_bytes + num_special + extra_merges

print(f"base byte tokens: {base_bytes}")
print(f"special tokens:  {num_special}")
print(f"extra merges:    {extra_merges}")
print(f"toy vocab size:  {toy_vocab_size}")
````

## Step 3. Train the toy tokenizer

This is the same call pattern the real repo uses, just on a much smaller corpus.

How this maps to `nanochat`:
- [scripts/tok_train.py](../scripts/tok_train.py) eventually calls `RustBPETokenizer.train_from_iterator(...)`.
- Inside that method, `rustbpe` learns merge rules from text.
- Then `nanochat` wraps the learned result in a `tiktoken.Encoding` for fast runtime encode/decode.


````python
from nanochat.tokenizer import RustBPETokenizer

if "toy_text_iterator" not in globals():
    raise RuntimeError("Run Step 1.5 first so `toy_text_iterator()` is defined.")
if "toy_vocab_size" not in globals():
    raise RuntimeError("Run Step 2 first so `toy_vocab_size` is defined.")

toy_tokenizer = RustBPETokenizer.train_from_iterator(toy_text_iterator(), vocab_size=toy_vocab_size)
print(f"Trained toy tokenizer with vocab size: {toy_tokenizer.get_vocab_size()}")
````

## Step 4. Look at the learned token pieces

The most important normal-text structure inside the tokenizer is `mergeable_ranks`.
It is the mapping from byte pieces to integer ranks/IDs.

How this maps to `nanochat`:
- `mergeable_ranks` is built inside `RustBPETokenizer.train_from_iterator(...)`.
- It is later passed into `tiktoken.Encoding(...)`.
- This is the learned vocabulary for ordinary text, separate from the chat special tokens.


````python
if "toy_tokenizer" not in globals():
    raise RuntimeError("Run Step 3 first so `toy_tokenizer` is defined.")

mergeable_ranks = getattr(toy_tokenizer.enc, "_mergeable_ranks")
interesting_pieces = [
    (rank, piece, piece.decode("utf-8", errors="replace"))
    for piece, rank in mergeable_ranks.items()
    if len(piece) > 1
]
interesting_pieces = sorted(interesting_pieces, key=lambda row: row[0])

print("A few learned non-trivial pieces from mergeable_ranks:")
print("rank | raw bytes | decoded text")
for rank, piece, decoded in interesting_pieces[:20]:
    print(f"{rank:>4} | {piece!r:<12} | {decoded!r}")
````

Notice that these pieces are **not** hand-coded English chunks.
They are just the byte sequences that were worth learning under the merge budget.

That is why a toy corpus may learn `b"lowest"` as one token, while a different corpus might stop at pieces like `b"low"` and `b"est"`.


## Step 5. Encode a few strings with the trained tokenizer

Now we use the tokenizer after training. At this point the tokenizer is fixed.
It is no longer learning anything; it is only applying the learned rules.

How this maps to `nanochat`:
- This is the mode used during actual model training and inference.
- The same saved tokenizer is reused again and again.


````python
if "toy_tokenizer" not in globals():
    raise RuntimeError("Run Step 3 first so `toy_tokenizer` is defined.")

def inspect_encoding(tokenizer, text):
    ids = tokenizer.encode(text)
    pieces = [tokenizer.decode([token_id]) for token_id in ids]
    return list(zip(ids, pieces))

for text in ["low", "lower", "lowest", "slowest", "brandnew"]:
    print(f"\n{text!r}")
    for token_id, piece in inspect_encoding(toy_tokenizer, text):
        print(f"  id={token_id:<4} piece={piece!r}")
````

`brandnew` is useful to inspect because it was not in the toy corpus as a full word.
The tokenizer still works by falling back to smaller learned pieces and raw byte tokens.

That is one of the big reasons BPE-style tokenizers are more flexible than simple word-level tokenizers.


## Step 6. Inspect nanochat special tokens

So far we only looked at ordinary text pieces.
`nanochat` also reserves special tokens used to structure chat data.

How this maps to `nanochat`:
- These tokens are what let the repo distinguish user text, assistant text, tool calls, and tool outputs.
- They are especially important for chat SFT and RL data formatting.


````python
from nanochat.tokenizer import SPECIAL_TOKENS

if "toy_tokenizer" not in globals():
    raise RuntimeError("Run Step 3 first so `toy_tokenizer` is defined.")

print("special token -> token id")
for token in SPECIAL_TOKENS:
    print(f"{token:<20} -> {toy_tokenizer.encode_special(token)}")
````

## Step 7. Render a chat conversation the way nanochat does

`RustBPETokenizer` is not only an encoder for plain strings.
It also knows how to turn a chat conversation into a single token sequence with special delimiters.

How this maps to `nanochat`:
- This is what happens before chat fine-tuning in [scripts/chat_sft.py](../scripts/chat_sft.py).
- `render_conversation(...)` also returns a `mask` so the training loss can focus on assistant tokens.


````python
if "toy_tokenizer" not in globals():
    raise RuntimeError("Run Step 3 first so `toy_tokenizer` is defined.")

conversation = {
    "messages": [
        {"role": "user", "content": "Why is the sky blue?"},
        {"role": "assistant", "content": "Short answer: scattering."},
    ]
}

ids, mask = toy_tokenizer.render_conversation(conversation)

print("token_id | piece | mask")
for token_id, mask_value in zip(ids, mask):
    piece = toy_tokenizer.decode([token_id])
    print(f"{token_id:>7} | {piece!r:<20} | {mask_value}")
````

A quick way to read that output:
- user wrapper tokens have `mask = 0`
- assistant wrapper tokens and assistant text have `mask = 1`
- this teaches the model to predict assistant responses, not the user prompt itself


## Step 8. See what `get_tokenizer()` does

`get_tokenizer()` is a loader, not a trainer.
It expects that `scripts/tok_train.py` has already created a tokenizer on disk.

How this maps to `nanochat`:
- `tok_train.py` writes the tokenizer to the base directory.
- `get_tokenizer()` loads it later for pretraining, chat training, eval, and inference.


````python
import inspect

import nanochat.tokenizer as tokenizer_module

print(inspect.getsource(tokenizer_module.get_tokenizer))

tokenizer_dir = Path(get_base_dir()) / "tokenizer"
print(f"Expected tokenizer directory: {tokenizer_dir}")
````

## Step 9.5. Turn tokenizer output into PyTorch tensors for `gpt.py`

Now we can look at the next boundary in `nanochat`:
- the tokenizer gives us token ids and a loss mask
- `scripts/chat_sft.py` turns those into `inputs` and `targets`
- `nanochat/gpt.py` receives those tensors as `idx` and `targets`

Important: the next two cells show a **toy version of the real repo flow**.
They do **not** train the model. They only build the final tensors that are ready to be passed into `gpt.py`.

The step-by-step shape of the conversion is:
1. `ids, mask = toy_tokenizer.render_conversation(conversation)` gives us `ids` and `mask` as plain Python lists.
2. `torch.tensor([ids])` turns one list of token ids into a 2D PyTorch tensor with a batch dimension.
3. We shift that tensor by one position to make `inputs` and `targets` for next-token prediction.
4. We use the mask to set non-assistant target positions to `-1`, which tells the loss to ignore them.

So the key conversion from tokenizer output to model-ready tensor is:
- Python token-id list: `ids`
- batched PyTorch tensor: `batch_tensor`
- final model input tensor: `gpt_ready_idx`
- final model target tensor: `gpt_ready_targets`

Which lines to pay attention to:
- in `scripts/chat_sft.py`: lines `286..297`, where `batch_tensor`, `inputs`, and `targets` are built
- in `nanochat/gpt.py`: lines `416..429`, where `forward(self, idx, targets=...)` starts and embeddings are looked up from `idx`


````python
print("chat_sft.py: building inputs and targets\n")
show_excerpt(repo_root / "scripts" / "chat_sft.py", 286, 297)
print("\n...\n")
print("gpt.py: model forward signature and first embedding lookup\n")
show_excerpt(repo_root / "nanochat" / "gpt.py", 416, 429)
````

````python
import torch

if "toy_tokenizer" not in globals() or "conversation" not in globals():
    raise RuntimeError("Run Step 3 and Step 7 first so `toy_tokenizer` and `conversation` are defined.")

# This is the exact tokenizer step from Step 7.
ids, mask = toy_tokenizer.render_conversation(conversation)

# At this point ids and mask are still ordinary Python lists produced by the tokenizer.
print("Before tensor conversion:")
print("ids:" + str(ids))
print("  ids, mask = toy_tokenizer.render_conversation(conversation)")
print(f"  type(ids)  = {type(ids)}")
print(f"  type(mask) = {type(mask)}")
print(f"  first 20 ids:  {ids[:20]}")
print(f"  first 20 mask: {mask[:20]}")
print()

# This mirrors the real chat_sft.py flow on a single toy conversation.
# Wrapping ids in [ids] adds a batch dimension, so shape becomes (1, sequence_length).
batch_tensor = torch.tensor([ids], dtype=torch.long)
mask_tensor = torch.tensor([mask], dtype=torch.int8)

# inputs are all tokens except the last one.
# targets are all tokens except the first one, so each target is the next token.
inputs = batch_tensor[:, :-1].to(dtype=torch.int32).contiguous()
targets = batch_tensor[:, 1:].to(dtype=torch.int64).contiguous()

# mask[1:] lines up with targets because targets are shifted by one token.
mask_targets = mask_tensor[:, 1:]
targets[mask_targets == 0] = -1

gpt_ready_idx = inputs
gpt_ready_targets = targets

print("After converting to tensors:")
print(f"  len(ids)  = {len(ids)}")
print(f"  len(mask) = {len(mask)}")
print(f"  batch_tensor.shape      = {tuple(batch_tensor.shape)}, dtype = {batch_tensor.dtype}")
print(f"  mask_tensor.shape       = {tuple(mask_tensor.shape)}, dtype = {mask_tensor.dtype}")
print(f"  gpt_ready_idx.shape     = {tuple(gpt_ready_idx.shape)}, dtype = {gpt_ready_idx.dtype}")
print(f"  gpt_ready_targets.shape = {tuple(gpt_ready_targets.shape)}, dtype = {gpt_ready_targets.dtype}")
print()
print("These tensors are ready to be passed into nanochat/gpt.py:")
print("  logits, loss = model(gpt_ready_idx, targets=gpt_ready_targets)")
print()
print("First 20 input ids:  ", gpt_ready_idx[0, :20].tolist())
print("First 20 target ids: ", gpt_ready_targets[0, :20].tolist())
ignored_positions = (gpt_ready_targets[0] == -1).nonzero(as_tuple=False).flatten().tolist()
print("Ignored target positions (mask == 0):", ignored_positions)
print()
print("Batch size is 1 here so the shapes stay simple. In real nanochat training,")
print("many rows like this are stacked together before calling the model.")
````

## Takeaways

- `RustBPETokenizer` learns a byte-based BPE vocabulary from text.
- After training, token pieces and token IDs are fixed.
- `mergeable_ranks` is the learned normal-text vocabulary.
- `nanochat` adds special tokens on top for chat formatting.
- `get_tokenizer()` loads the saved tokenizer; it does not retrain it.
- `tok_train.py` creates the tokenizer, and `tok_eval.py` measures how good its compression is.
